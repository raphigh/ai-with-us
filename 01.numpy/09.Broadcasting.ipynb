```python
import numpy as np
np.random.seed(1400)
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در کد زیر، با استفاده از خاصیت پخش آرایه در نامپای، دو آرایه a و b را در هم ضرب می‌کنیم و نتیجه، ضرب نظیر به نظیر عناصر آرایه‌ها می‌باشد.
</font>
</p>


```python
a = np.array([0, 1, 2, 3])
b = np.array([2, 2, 2, 2])
a * b
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در نامپای می‌توانیم با ضرب کردن یک عدد در آرایه هم به همین نتیجه برسیم.
</font>
</p>



```python
2 * a
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در کد زیر، آرایه b در محور اول کش آمده است.</font>
</p>


```python
a = np.arange(12).reshape(3, 4)
print(a)
b = np.array([3, 2, 1, 0])
print(b)
a + b
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در کد زیر، آرایه b در محور دوم کش آمده است.</font>
</p>


```python
b = np.array([8, 4, 0]).reshape(3, 1)
print(b)
a + b
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
این‌ بار هر دو آرایه کش آمده‌اند و همشکل شده‌اند.</font>
</p>


```python
a = np.array([1, 2, 3, 4, 5])
b = np.array([1, 2, 3, 4]).reshape(4, 1)
print(a)
print(b)
a * b
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
یک آرایه دو بعدی شامل اعداد تصادفی یک تا پنج ایجاد می‌کنیم.</font>
</p>


```python
a = np.random.normal(5, 1, (4, 5))
print(a)
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
میانگین هر ستون را در آرایه <code>col_mean</code> میریزیم.</font>
</p>


```python
col_mean = a.mean(axis=0)
print(col_mean)
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
با کمک گرفتن از ویژگی پخش، میانگین هر ستون را از تک تک عناصر آن ستون کم می‌کنیم.
</p>


```python
a -= col_mean
print(a)
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
بار دیگر میانگین هر سطر را محاسبه می‌کنیم که می‌بینید برای هر ستون تقریبا صفر است.
</p>


```python
a.mean(axis=0)
```
